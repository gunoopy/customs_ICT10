## 변수의 이름 규칙
- 변수 이름은 문자, 숫자, 밑줄(_)로 구성될 수 있다.
- 변수 이름은 숫자로 시작할 수 없다.
- 변수 이름은 대소문자를 구분한다.
- 예약어(파이썬에서 미리 정의된 키워드)는 변수 이름으로 사용할 수 없다.
- 변수 이름은 변수의 의미나 용도를 나타내도록 짓는 것이 좋다.

```python
# 권장함
student_name = "John"
total_score = 95

# 권장하지 않음
a = "John"
b = 95
```
