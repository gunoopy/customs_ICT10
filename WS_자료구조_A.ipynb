### Q2) 당신은 치킨집 사장님입니다. 밀려드는 주문을 파이썬으로 관리해보세요.

- 현재 배달 주문 목록을 만드세요.
    - orders에  "치킨", "피자", "치킨", "떡볶이", "치킨", "피자", "순대", "치킨" 주문을 리스트로 만드세요.

```python
orders = ["치킨", "피자", "치킨", "떡볶이", "치킨", "피자", "순대", "치킨"]
orders
```

- 첫 번째로 들어온 주문과 가장 마지막 주문이 뭔지 확인하세요.

```python
first_order = orders[0]
last_order = orders[-1]
print(f"첫 주문: {first_order}, 마지막 주문: {last_order}")
```

- 주방 직원이 딱 3명뿐입니다. 주문 중 앞에서 3개만 먼저 처리하고, 나머지 주문은 대기열로 따로 빼두세요.

```python
processing = orders[:3]
waiting = orders[3:]
print(f"처리 중: {processing}")
print(f"대기 중: {waiting}")
```

- 총 주문 수를 구하고, 주방이 바쁜지 판단하세요.
    - 주문이 6개 초과면 "🔥 주방 풀가동!", 이하면 "😊 여유 있음" 을 출력하세요.

```python
total = len(orders)

if total > 6:
    print("🔥 주방 풀가동!")
else:
    print("😊 여유 있음")
```

- 새 주문이 들어왔습니다. "마라탕"과 "탕수육"을 순서대로 주문 목록에 추가하세요.

```python
orders.append("마라탕")
orders.append("탕수육")
orders
```

- 갑자기 "순대" 주문이 취소됐습니다. 목록에서 지워주세요.

```python
orders.remove("순대")
orders
```

- 오늘 치킨이 몇 개나 팔렸는지 세어보고, 인기 메뉴인지 확인하세요.
    - 3개 이상이면 "🏆 오늘의 인기 메뉴!", 미만이면 "📉 판매 부진" 을 출력하세요.

```python
chicken_count = orders.count("치킨")

if chicken_count >= 3:
    print(f"치킨 {chicken_count}개 판매 — 🏆 오늘의 인기 메뉴!")
else:
    print(f"치킨 {chicken_count}개 판매 — 📉 판매 부진")
```
