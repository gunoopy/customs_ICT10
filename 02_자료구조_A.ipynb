# 자료구조 - 리스트
- 리스트(list)는 가장 많이 사용되는 내장 자료구조 중 하나로, 순서가 있는 변경 가능한 객체들의 집합이다.
- 리스트는 다양한 데이터 타입을 저장할 수 있으며, 동적으로 크기를 조절할 수 있다.
- 리스트에 대한 주요 특징과 사용법은 다음과 같다.
    - 순서가 있는 컬렉션 : 리스트에 저장된 요소들은 순서가 있으며, 인덱스를 통해 접근할 수 있다. 인덱스는 0부터 시작한다.
    - 다양한 데이터 타입 저장 가능: 리스트는 정수, 문자열, 다른 리스트 등 다양한 데이터 타입을 저장할 수 있다.
    - 동적 크기 조절: 리스트의 크기는 요소를 추가하거나 제거함에 따라 동적으로 변경된다
    - 변경 가능: 리스트는 요소를 추가, 삭제, 수정할 수 있다.
- 리스트는 대괄호 []를 사용하여 생성할 수 있다.
    - list() 함수를 사용하여 다른 자료구조를 리스트 자료구조로 변환할 수 있다.
- 리스트의 요소는 인덱스를 사용하여 접근할 수 있으며, 인덱스를 통해 요소를 수정할 수도 있다.
- 리스트는 다양한 내장 함수를 제공하여 요소를 추가, 삭제, 검색하는 등의 작업을 쉽게 수행할 수 있다.

```python
# 리스트 생성
numbers = [1, 2, 3, 4, 5]
numbers
```

```python
# 리스트 인덱싱 - 첫번째 값 선택
numbers[0]
```

```python
# 리스트 인덱싱 - 마지막 값 선택
numbers[-1]
```

```python
# 리스트 값 수정 - 첫번째 값을 10으로 수정
numbers[0] = 10
numbers
```

```python
# append(x): 리스트의 끝에 요소 x를 추가
my_list = [10, 20, 30]
my_list.append(40)
my_list
```

```python
# remove(x): 리스트에서 첫 번째로 나타나는 요소 x를 제거
my_list = [10, 20, 30, 20, 40]
my_list.remove(20)
my_list
```

```python
# count(x): 리스트에서 요소 x의 개수를 반환
my_list = [10, 20, 20, 30, 20, 40]
my_list.count(20)
```

```python
# sort(key=None, reverse=False): 리스트를 정렬
my_list = [3, 1, 4, 2]
my_list.sort()
my_list
```

```python
my_list.sort(reverse=True)
my_list
```

# 자료구조 - 딕셔너리
- 딕셔너리(dict)는 기본 자료 구조 중 하나로, 키(key)와 값(value)의 쌍을 저장하는 데이터 구조이다.
- 딕셔너리는 JSON(JavaScript Object Notation), 다른 언어의 해시맵(HashMap)이나 연관 배열(Associative Array)과 유사하다.
- 세트에 대한 주요 특징과 사용법은 다음과 같다.
    - 키-값 쌍: 딕셔너리는 각 키가 고유하며, 이 키에 대응하는 값이 저장된다.
    - 중복 키 허용 안 함: 딕셔너리의 키는 중복될 수 없다. 동일한 키로 값을 저장하면 기존 값이 덮어쓰여진다.
    - 순서 없음: Python 3.6 이후부터는 삽입 순서를 유지하지만, 기본적으로는 순서가 보장되지 않는 자료구조이다.
    - 변경 가능 : 딕셔너리는 요소를 추가, 변경, 삭제할 수 있다.
- 딕셔너리는 중괄호 {}와 콜론(:) 을 사용하여 생성할 수 있다.
    - dict() 함수를 사용하여 딕셔너리 자료구조를 생성할 수 있다.
- 딕셔너리 요소에 접근하거나 변경할 때는 키를 사용한다.


```python
# 딕셔너리 생성
student = {"name": "John", "age": 25, "major": "Computer Science"}
student
```

```python
# "name" 키 값으로 "John" 값에 접근
student["name"]
```

```python
# "age" 키의 값을 26으로 변경
student["age"] = 26
student
```

```python
# "GPA" : 3.8 쌍 추가
student["GPA"] = 3.8
student
```

```python
student = {"name": "John", "age": 25, "major": "Computer Science"}

# 모든 키를 가져오기
print(student.keys())  # 출력: dict_keys(['name', 'age', 'major'])

# 모든 값을 가져오기
print(student.values())  # 출력: dict_values(['John', 25, 'Computer Science'])

# 모든 키-값 쌍을 가져오기
print(student.items())  # 출력: dict_items([('name', 'John'), ('age', 25), ('major', 'Computer Science')])
```
