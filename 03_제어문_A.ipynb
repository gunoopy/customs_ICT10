# 제어문
- 제어문은 프로그램의 흐름을 제어하는 데 사용되는 명령문으로
- 제어문을 통해 프로그램은 순차적 실행에서 벗어나 조건에 따라 분기하거나 특정 코드 블록을 반복 실행할 수 있다.
- 제어문을 사용하면 코드의 재사용성과 유지 보수성을 높일 수 있다.
- Python에서 자주 사용하는 제어문은 조건문(Conditional Statements)과 반복문(Loop Statements)이 존재한다.

# 조건문
- 조건문은 특정 조건에 따라 코드 블록을 실행하거나 실행하지 않도록 제어하는 데 사용된다.
- 조건문을 사용하면 프로그램이 다양한 상황에 적절히 대응할 수 있도록 유연하게 코드를 작성할 수 있다.
- Python에서 사용되는 주요 조건문은 if, elif, else가 있다.
    - if 문은 주어진 조건식이 참(True)일 때만 코드 블록을 실행한다.
    - elif 문은 앞선 if 문이나 다른 elif 문 조건이 거짓(False)일 때, 새로운 조건식을 평가하여 참일 경우 코드 블록을 실행한다.
    - else 문은 모든 조건식이 거짓(False)일 때 실행된다.
    - elif 문은 여러 개 사용할 수 있으며 else 문은 if 또는 elif 문 뒤에 하나만 사용할 수 있다.

```python
age = 22

if age >= 20 :
    print("성인입니다.")
    print("결제 금액 : 15000원")
else :
    print("청소년입니다.")
    print("결제 금액 : 12000원")

print("결제 완료")
```

```python
age = 18

if age >= 65 :
    print("결제 금액 : 10000원")
elif age >= 20 :
    print("결제 금액 : 15000원")
elif age >= 14 :
    print("결제 금액 : 12000원")
else :
    print("무료입니다.")

print("결제 완료")
```

- 조건(condition) 자리에 많이 사용되는 연산자
    - 비교 연산자 : < > <= >= == !=
    - 논리 연산자 : or / and / not
    - 멤버쉽 연산자 : in / not in

```python
today = '목요일'

# today가 토요일 이거나 일요일이면 "주말입니다." 출력, 그렇지 않으면 "평일입니다." 출력
if today == '토요일' or today == '일요일' :
    print('주말입니다.')
else :
    print('평일입니다.')
```

# [실습] 조건문

- score의 값에 따라 다음과 같이 성적을 출력하는 조건문을 작성하세요:
    - 90 이상이면 "A"
    - 80 이상 90 미만이면 "B"
    - 70 이상 80 미만이면 "C"
    - 60 이상 70 미만이면 "D"
    - 60 미만이면 "F"

```python
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print("Grade:", grade)
```

- 나이와 키 정보를 입력받아 다음 조건에 맞는 문구를 출력하세요.
    - 나이가 5살 미만이거나 키가 110cm 미만이면 → "보호자 동반 필수 구역입니다."
    - 나이가 13살 이상이고 키가 140cm 이상이면 → "모든 놀이기구 탑승 가능합니다!"
    - 그 외 → "회전목마, 범퍼카 등 순한 놀이기구만 탑승 가능합니다."

```python
age = int(input("나이를 입력하세요: "))
height = int(input("키를 입력하세요 (cm): "))

if age < 5 or height < 110:
    print("보호자 동반 필수 구역입니다.")
elif age >= 13 and height >= 140:
    print("모든 놀이기구 탑승 가능합니다!")
else:
    print("회전목마, 범퍼카 등 순한 놀이기구만 탑승 가능합니다.")
```

# 반복문 - for
- Python의 for 반복문은 주어진 시퀀스(리스트, 튜플, 문자열 등)나 다른 반복 가능한 객체(iterable)의 각 요소를 순회하면서 코드 블록을 반복 실행한다. (사전에 반복 횟수가 정해진  경우에 유용)
- 변수는 시퀀스의 각 요소를 순차적으로 받아들이는 변수이며, 시퀀스는 리스트, 튜플, 문자열 또는 다른 반복 가능한 객체를 의미한다.
- range() 함수는 정수 범위를 생성한다.
    - range(start, stop, step) 형식으로 사용되며, 각 매개변수는 다음을 의미한다.
    - start: 시작 값 (기본값은 0)
    - stop: 끝 값 (포함되지 않음)
    - step: 증가 값 (기본값은 1)

```python
# 리스트의 각 요소를 출력
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit.upper())
    print(fruit[:3])
```

```python
# 0부터 4까지 숫자를 출력
for i in range(5):
    print(i)
```

```python
# 0부터 4까지 제곱 수 출력
for i in range(5):
    print(i**2)
```

# [실습] 반복문

- range 함수를 사용해서 원하는 단의 구구단을 출력하세요.
    - 문자열 포맷팅 : f"{값1} x {값2} = {값3}"

```python
dan = 5

for i in range(1, 10):
    print(f"{dan} × {i} = {dan * i}")
```

- 다음 점수 리스트를 반복문을 돌면서 점수를 하나씩 더해 합계를 구하세요.

```python
scores = [78, 92, 65, 88, 73, 95, 60]

total = 0

for score in scores:
    total = total + score

total
```
